**PANDAS - Data handling & Preprocessing**

```python
import pandas as pd
```

```python
data={
    'Age':[25,30,35,40],
    'Salary':[30000,50000,70000,90000],
    'Department':['IT','HR','IT','Finance']
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.shape)
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df['Salary'])
print(df[['Age','Department']])
```

```python
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
```

```python
df_encoded=pd.get_dummies(df,drop_first=True)
print(df_encoded)
```

**MATPLOTLIB**

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,25,30]
```

```python
plt.plot(x,y)
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.title('Line Plot')
plt.show()
```

```python
plt.bar(x,y)
plt.title('Bar Plot')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.show()
```

```python
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.show()
```

**SEABORN**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#LOAD SAMPLE DATASET
data=sns.load_dataset("iris")
```

```python
sns.pairplot(data,hue='species')
plt.show()
```

```python
corr=data.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_length',data=data)
plt.show()
```
